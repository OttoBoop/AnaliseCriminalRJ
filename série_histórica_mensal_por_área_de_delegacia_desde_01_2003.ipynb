```python
import pandas as pd

# URL do arquivo bruto (raw) no GitHub
file_url = "https://raw.githubusercontent.com/OttoBoop/AnaliseCriminalRJ/main/BaseDPEvolucaoMensalCisp.csv"

# Carregar o CSV diretamente do GitHub
df = pd.read_csv(file_url, sep=";", encoding="latin1")

# Exibir as primeiras linhas para conferir a importação
print(df.head())
```

```python
import matplotlib.pyplot as plt

# Filtrar apenas para o ano de 2003 e agrupar por mês
meses = range(1, 13)
valores = []

for mes in meses:
    if not df[(df["ano"] == 2003) & (df["mes"] == mes)]["roubo_celular"].empty:
        valores.append(df[(df["ano"] == 2003) & (df["mes"] == mes)]["roubo_celular"].sum())
    else:
        valores.append(0)

# Criar o gráfico
plt.figure(figsize=(10, 5))
plt.plot(meses, valores, marker='o', linestyle='-', color='b', label="Roubos de Celular")
plt.xlabel("Mês")
plt.ylabel("Quantidade")
plt.title("Roubos de Celular por Mês - 2003")
plt.xticks(meses)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

anos = range(2003, 2025)
valores = []

for ano in anos:
    if not df[df["ano"] == ano]["roubo_celular"].empty:
        valores.append(df[df["ano"] == ano]["roubo_celular"].sum())
    else:
        valores.append(0)

plt.figure(figsize=(12, 6))
plt.plot(anos, valores, marker='o', linestyle='-', color='b', label="Total de Roubos de Celular")

plt.xlabel("Ano")
plt.ylabel("Quantidade")
plt.title("Total de Roubos de Celular por Ano (2003-2024)")
plt.xticks(anos, rotation=45)
plt.grid(True)
plt.legend()
plt.show()
```

```python
crime_columns = [
    'hom_doloso', 'lesao_corp_morte', 'latrocinio', 'cvli',
    'hom_por_interv_policial', 'letalidade_violenta', 'tentat_hom',
    'lesao_corp_dolosa', 'estupro', 'hom_culposo', 'lesao_corp_culposa',
    'roubo_transeunte', 'roubo_celular', 'roubo_em_coletivo', 'roubo_rua',
    'roubo_veiculo', 'roubo_carga', 'roubo_comercio', 'roubo_residencia',
    'roubo_banco', 'roubo_cx_eletronico', 'roubo_conducao_saque',
    'roubo_apos_saque', 'roubo_bicicleta', 'outros_roubos', 'total_roubos',
    'furto_veiculos', 'furto_transeunte', 'furto_coletivo', 'furto_celular',
    'furto_bicicleta', 'outros_furtos', 'total_furtos', 'sequestro',
    'extorsao', 'sequestro_relampago', 'estelionato', 'apreensao_drogas',
    'posse_drogas', 'trafico_drogas', 'apreensao_drogas_sem_autor',
    'recuperacao_veiculos', 'apf', 'aaapai', 'cmp', 'cmba', 'ameaca',
    'pessoas_desaparecidas', 'encontro_cadaver', 'encontro_ossada',
    'pol_militares_mortos_serv', 'pol_civis_mortos_serv'
]
```

```python
# Criar um novo DataFrame com a soma de cada crime por ano
df_agrupado = df.groupby("ano")[crime_columns].sum().reset_index()
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go

# Inicializar o app
app = dash.Dash(__name__)

# Layout do app
app.layout = html.Div([
    html.H1("Evolução dos Crimes no Rio de Janeiro (2003-2024)"),

    # Dropdown para selecionar o crime
    dcc.Dropdown(
        id='crime-dropdown',
        options=[{'label': crime, 'value': crime} for crime in crime_columns],
        value='roubo_celular',  # Valor padrão
        clearable=False
    ),

    # Gráfico interativo
    dcc.Graph(id='crime-graph')
])

# Callback para atualizar o gráfico com base no crime escolhido
@app.callback(
    Output('crime-graph', 'figure'),
    Input('crime-dropdown', 'value')
)
def update_graph(selected_crime):
    fig = go.Figure()

    # Adicionar a linha com pontos nos anos
    fig.add_trace(go.Scatter(
        x=df_agrupado["ano"],
        y=df_agrupado[selected_crime],
        mode="lines+markers",  # Linha conectando os pontos e marcadores nos anos
        marker=dict(size=8, color='blue'),  # Ajustar tamanho e cor dos pontos
        line=dict(color='blue', width=2),  # Ajustar a linha
        name=selected_crime
    ))

    # Personalizar o layout
    fig.update_layout(
        title=f"Evolução de {selected_crime} (2003-2024)",
        xaxis_title="Ano",
        yaxis_title="Quantidade",
        plot_bgcolor="white",  # Fundo branco
        paper_bgcolor="white",  # Fundo externo branco
        xaxis=dict(showgrid=True, gridcolor='lightgray'),  # Adicionar grid
        yaxis=dict(showgrid=True, gridcolor='lightgray')
    )

    return fig

# Rodar o app
if __name__ == '__main__':
    app.run_server(debug=True, port=8051)
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go

# Inicializar o app
app = dash.Dash(__name__)

# Layout do app
app.layout = html.Div([
    html.H1("Evolução dos Crimes no Rio de Janeiro (2003-2024)"),

    # Dropdown para selecionar o crime
    dcc.Dropdown(
        id='crime-dropdown',
        options=[{'label': crime, 'value': crime} for crime in crime_columns],
        value='roubo_celular',
        clearable=False
    ),

    # Toggle de granularidade
    dcc.RadioItems(
        id="granularidade",
        options=[
            {'label': 'Anual', 'value': 'anual'},
            {'label': 'Mensal', 'value': 'mensal'}
        ],
        value='anual',
        inline=True
    ),

    # Gráfico interativo
    dcc.Graph(id='crime-graph')
])

# Callback para atualizar o gráfico
@app.callback(
    Output('crime-graph', 'figure'),
    [Input('crime-dropdown', 'value'),
     Input('granularidade', 'value')]
)
def update_graph(selected_crime, granularidade):
    fig = go.Figure()

    if granularidade == 'anual':
        # Gráfico por ano
        fig.add_trace(go.Scatter(
            x=df_anual["ano"],
            y=df_anual[selected_crime],
            mode="lines+markers",
            marker=dict(size=8, color='blue'),
            line=dict(color='blue', width=2),
            name=selected_crime
        ))

        fig.update_layout(
            title=f"Evolução de {selected_crime} (2003-2024) - Anual",
            xaxis_title="Ano",
            yaxis_title="Quantidade",
            plot_bgcolor="white",
            paper_bgcolor="white",
            xaxis=dict(showgrid=True, gridcolor='lightgray'),
            yaxis=dict(showgrid=True, gridcolor='lightgray')
        )

    else:
        # Gráfico por mês
        fig.add_trace(go.Scatter(
            x=df_mensal["data"],
            y=df_mensal[selected_crime],
            mode="lines+markers",
            marker=dict(size=4, color='blue'),
            line=dict(color='blue', width=1.5),
            name=selected_crime
        ))

        # Adicionar gridlines verticais em Janeiro de cada ano
        for ano in range(2003, 2025):
            fig.add_vline(x=f"{ano}-01", line_width=1, line_dash="dash", line_color="gray")

        fig.update_layout(
            title=f"Evolução de {selected_crime} (2003-2024) - Mensal",
            xaxis_title="Mês",
            yaxis_title="Quantidade",
            plot_bgcolor="white",
            paper_bgcolor="white",
            xaxis=dict(showgrid=True, gridcolor='lightgray'),
            yaxis=dict(showgrid=True, gridcolor='lightgray'),
            xaxis_tickangle=-45
        )

    return fig

# Rodar o app
if __name__ == '__main__':
    app.run_server(debug=True, port=8052)

```

```python
# Calcular a média dos crimes entre 2003-2005 e 2022-2024
media_2003_2005 = df_rj[df_rj["ano"].between(2003, 2005)][crime_columns].mean()
media_2022_2024 = df_rj[df_rj["ano"].between(2022, 2024)][crime_columns].mean()

# Calcular a variação percentual entre os períodos
variacao_percentual = ((media_2022_2024 - media_2003_2005) / media_2003_2005) * 100

# Criar um DataFrame ordenado por variação percentual
df_variacao = pd.DataFrame({
    "Média 2003-2005": media_2003_2005,
    "Média 2022-2024": media_2022_2024,
    "Variação (%)": variacao_percentual
}).sort_values(by="Variação (%)", ascending=False)
```

```python
df_homicidios = df_rj.groupby("ano").agg({
    "hom_doloso": "sum",
    "lesao_corp_morte": "sum",
    "latrocinio": "sum",
    "cvli": "sum",
    "hom_por_interv_policial": "sum",
    "pol_militares_mortos_serv": "sum",
    "pol_civis_mortos_serv": "sum",
    "hom_culposo": "sum",
    "pessoas_desaparecidas": "sum",
    "encontro_cadaver": "sum",
    "encontro_ossada": "sum"
}).reset_index()

# Criar as categorias agregadas
df_homicidios["Homicídios Violentos Gerais (Sem Polícia)"] = (
    df_homicidios["hom_doloso"] + df_homicidios["lesao_corp_morte"] +
    df_homicidios["latrocinio"] + df_homicidios["cvli"]
)

df_homicidios["Homicídios Violentos Gerais (Com Polícia)"] = (
    df_homicidios["Homicídios Violentos Gerais (Sem Polícia)"] + df_homicidios["hom_por_interv_policial"]
)

df_homicidios["Homicídios de Policiais em Serviço"] = (
    df_homicidios["pol_militares_mortos_serv"] + df_homicidios["pol_civis_mortos_serv"]
)

df_homicidios["Toda Letalidade Confirmada"] = (
    df_homicidios["Homicídios Violentos Gerais (Com Polícia)"] +
    df_homicidios["Homicídios de Policiais em Serviço"] +
    df_homicidios["hom_culposo"]
)

df_homicidios["Casos Suspeitos"] = (
    df_homicidios["pessoas_desaparecidas"] +
    df_homicidios["encontro_cadaver"] +
    df_homicidios["encontro_ossada"]
)

df_homicidios["Toda Letalidade Confirmada + Casos Suspeitos"] = (
    df_homicidios["Toda Letalidade Confirmada"] + df_homicidios["Casos Suspeitos"]
)
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.graph_objects as go

# Inicializar o app
app = dash.Dash(__name__)

# Layout do app
app.layout = html.Div([
    html.H1("Evolução dos Homicídios e Desaparecimentos (2003-2024)"),

    # Gráfico interativo
    dcc.Graph(id='homicidios-graph')
])

# Callback para atualizar o gráfico
@app.callback(
    Output('homicidios-graph', 'figure'),
    Input('homicidios-graph', 'id')  # Apenas para disparar o callback
)
def update_graph(_):
    fig = go.Figure()

    # Adicionar todas as séries de dados
    for col in [
        "Homicídios Violentos Gerais (Sem Polícia)",
        "Homicídios Violentos Gerais (Com Polícia)",
        "Homicídios de Policiais em Serviço",
        "Toda Letalidade Confirmada",
        "Casos Suspeitos",
        "Toda Letalidade Confirmada + Casos Suspeitos"
    ]:
        fig.add_trace(go.Scatter(
            x=df_homicidios["ano"],
            y=df_homicidios[col],
            mode="lines+markers",
            name=col
        ))

    # Configuração do layout
    fig.update_layout(
        title="Evolução dos Homicídios e Desaparecimentos (2003-2024)",
        xaxis_title="Ano",
        yaxis_title="Quantidade",
        plot_bgcolor="white",
        paper_bgcolor="white",
        xaxis=dict(showgrid=True, gridcolor='lightgray'),
        yaxis=dict(showgrid=True, gridcolor='lightgray'),
        legend_title="Categorias"
    )

    return fig

# Rodar o app
if __name__ == '__main__':
    app.run_server(debug=True, port=8053)
```

```python

df_homicidios["% Letalidade Confirmada"] = (
    df_homicidios["Toda Letalidade Confirmada"] / df_homicidios["Toda Letalidade Confirmada + Casos Suspeitos"]
) * 100
# Selecionar apenas as colunas relevantes para a análise
df_percentual = df_homicidios[[
    "ano",
    "Toda Letalidade Confirmada",
    "Casos Suspeitos",
    "Toda Letalidade Confirmada + Casos Suspeitos",
    "% Letalidade Confirmada"
]]
```

```python
import plotly.graph_objects as go

# Criar o gráfico
fig = go.Figure()

# Adicionar as séries de dados
for col in [
    "Toda Letalidade Confirmada",
    "Casos Suspeitos",
    "Toda Letalidade Confirmada + Casos Suspeitos"
]:
    fig.add_trace(go.Scatter(
        x=df_homicidios["ano"],
        y=df_homicidios[col],
        mode="lines+markers",
        name=col
    ))

# Adicionar a nova linha de percentual
fig.add_trace(go.Scatter(
    x=df_homicidios["ano"],
    y=df_homicidios["% Letalidade Confirmada"],
    mode="lines+markers",
    name="% Letalidade Confirmada",
    yaxis="y2",
    line=dict(dash="dot", color="purple")
))

# Configurar eixos duplos (letalidade normal e porcentagem)
fig.update_layout(
    title="Evolução da Letalidade Confirmada e Casos Suspeitos (2003-2024)",
    xaxis_title="Ano",
    yaxis=dict(title="Quantidade Absoluta", showgrid=True, gridcolor='lightgray'),
    yaxis2=dict(
        title="Porcentagem (%)",
        overlaying="y",
        side="right",
        showgrid=False
    ),
    plot_bgcolor="white",
    paper_bgcolor="white",
    legend_title="Categorias"
)

# Exibir o gráfico
fig.show()
```
